```python
def compressString(string):
    n = len(string)
    string = list(string.rstrip())
    i = 0
    count = 0
    for j in range(n):
        
        if string[i] != string[j]:
            i += 1
            string[i] = string[j]
            if count > 1:
                string[i] = str(count)
                i += 1
            count = 1
        else:
            count += 1
            
              
    if count > 1:
        string[i] = string[j]
        string[i] = str(count)
        i += 1
    
#     i+=1
    string = ''.join(string[:i])
    return string
```

```python
def compressString(string):
    n = len(string)
    res = ""
    count = 1
    for i in range(1,n):
        
        if string[i-1] != string[i]:
            res += string[i-1]
            if count > 1:
                res += str(count)
            count = 1
        else:
            count += 1
            
    res += string[-1]
    if count > 1:
        res += str(count)
    return res
```
