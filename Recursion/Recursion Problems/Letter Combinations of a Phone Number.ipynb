```python
def letterCombinations(digits, p=''):
    lst = []
    if digits == '':
        return [p]
    
    digit = int(digits[0])
    i = (digit-2)*3
    while i < (digit-1) * 3:
        ch = chr(ord('a')+i)
        lst += letterCombinations(digits[1:], p+ch)
        i += 1
     
    return lst
```
