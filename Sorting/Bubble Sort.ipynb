```python
def bubbleSortRecursive(arr, n, row=0):
    if n <= 1:
        return arr
    if row > n-1:
        return bubbleSortRecursive(arr, n - 1, 0)
    if arr[row] > arr[n-1]:
        arr[row], arr[n-1] = arr[n-1], arr[row]
    return bubbleSortRecursive(arr, n, row + 1)
```

```python
def bubbleSort(arr, n):
    for i in range(n):
        for j in range(i+1, n):
            if arr[i] > arr[j]:
                arr[i], arr[j] = arr[j], arr[i]
    return arr
            
```

TC :    
worst case : O(n^2)   
     best case : O(n)
