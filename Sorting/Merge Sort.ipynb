```python
def mergeSort(arr,n=0):

    def merge(a, b, res=[]):
        if not a and not b:
            return res
        if not a:
            res.extend(b)
            return res
        if not b:
            res.extend(a)
            return res
        if a[0] < b[0]:
            res.append(a[0])
            return merge(a[1:], b, res)
        else:
            res.append(b[0])
            return merge(a, b[1:], res)

    if not arr:
        return arr
    if len(arr) == 1:
        return arr
    mid = len(arr) // 2
    s1 = mergeSort(arr[:mid])
    s2 = mergeSort(arr[mid:])
    return merge(s1, s2)
```

worst case : O(nlongn)   
     best case : O(n)
