```python
def isHappyNum(n):
    def findSquare(n):
        ans = 0
        while n > 0:
            rev = n%10
            ans +=  rev * rev
            n = n//10
        return ans
    f =s = n        
    s = findSquare(s)
    f = findSquare(findSquare(f))
    while s != f:
        s = findSquare(s)
        f = findSquare(findSquare(f))
    if s == 1:
        return True
    return False
        
```
