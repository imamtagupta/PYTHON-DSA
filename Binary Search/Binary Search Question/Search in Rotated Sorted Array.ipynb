```python
def binarySearch(s,e,t,nums):
    while s <= e:
        mid = (s+e)//2
        if nums[mid] == t:
            return mid
        elif nums[mid] > t:
            e = mid - 1
        else:
            s = mid + 1
    return -1

def findPivot(nums):
    n = len(nums)
    s = 0
    e = n-1
    if n==0:
        return -1
    if n==1:
        return 0
    while e-s > 1:
        mid = s+((e-s)//2)
        if mid < e and nums[mid] > nums[mid+1]:
            return mid
        elif mid > s and nums[mid-1] > nums[mid]:
            return mid-1
        elif nums[s] >= nums[mid]:
            e = mid - 1
        else:
            s = mid + 1
    return s if nums[s]>nums[e] else e

def searchInRotatedSortedArray(nums, target):
    n = len(nums)
    p = findPivot(nums)
    if nums[p] == target:
        return p
    elif nums[0] <= target:
        return binarySearch(0, p-1, target, nums)
    else:
        return binarySearch(p+1, n-1, target, nums)

    
        
```
