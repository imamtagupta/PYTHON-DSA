### Here a target is given and we suppose to return its ceiling ie smallest number n >= target

```python
def ceilingOfArray(nums, target):
    low = 0
    high = len(nums)-1
    if len(nums) > 0 and nums[high] >= target:
        while low <= high:
            mid = (low+high) // 2
            if nums[mid] == target:
                return nums[mid]
            elif nums[mid] > target:
                high = mid - 1
            else:
                low = mid + 1
        return nums[low]
    return -1
```
