### Here a target is given and we suppose to return its floor ie greatest number n <= target

```python
def flooringOfArray(nums, target):
    low = 0
    high = len(nums)-1
    if len(nums) > 0 and nums[low] <= target:
        while low <= high:
            mid = (low+high) // 2
            if nums[mid] == target:
                return nums[mid]
            elif nums[mid] > target:
                high = mid - 1
            else:
                low = mid + 1
        return nums[high]
    return -1
```
