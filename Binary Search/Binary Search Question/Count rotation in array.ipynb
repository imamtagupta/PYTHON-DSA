```python
def countRotationOfArray(nums):
    s = 0
    n = len(nums)
    e = n - 1
    while s<=e:
        mid = s + ((e-s) // 2)
        if mid > 0 and nums[mid] < nums[mid-1]:
            return mid
        if mid < n-1 and nums[mid] > nums[mid+1]:
            return mid + 1
        if nums[mid] > nums[mid-1] and nums[mid] <= nums[0]:
            e = mid - 1
        else:
            s = mid + 1
    return 0
```
