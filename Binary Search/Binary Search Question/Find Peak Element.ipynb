```python
def peakElementArray(nums):
    n = len(nums)
    lo = 0
    hi = n-1
    if n<1:
        return -1
    elif n==1:
        return lo
    while hi-lo>1:
        mid = lo + ((hi - lo)//2)
        prev = (n + (mid - 1)) % n
        nxt = (mid + 1) % n
        if nums[prev] < nums[mid] > nums[nxt]:
            return mid
        elif nums[prev] > nums[mid]:
            hi = mid - 1
        else:
            lo = mid + 1
    return lo if nums[lo] > nums[hi] else hi
        
```
