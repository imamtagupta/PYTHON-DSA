```python
def binarySearch(s,e,t,nums):
    while e >= s:
        mid = (s+e)//2
        print(mid)
        if nums[mid] == t:
            return mid
        elif nums[mid] > t:
            e = mid - 1
        else:
            s = mid + 1
    return -1


def findInfiniteArray(nums, target):
    s=0
    e=s+1
    while target>nums[e]:
        step = e+(e-s+1)*2
        print(step)
        s,e = e+1,step
    return binarySearch(s,e,target,nums)
```
