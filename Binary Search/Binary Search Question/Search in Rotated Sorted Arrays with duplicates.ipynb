```python
def binarySearch(s,e,nums,t):
    print(s,e,nums,t)
    if e-s < 1:
        if nums[s] == t:
            return True
        return False
        
    while s <= e:
        mid = (s+e)//2
        if nums[mid] == t:
            return True
        elif nums[mid] > t:
            e = mid - 1
        else:
            s = mid + 1
    return False

def findPivot(nums):
    n = len(nums)
    s = 0
    e = n-1
    mid = s + ((e-s) // 2)
    if nums[s] == nums[e]:
        s += 1
        e -= 1
    while s<=e:
        if mid < e and nums[mid] > nums[mid+1]:
            return mid
        elif mid > s and nums[mid-1] > nums[mid]:
            return mid-1
        elif nums[mid] <= nums[s]:
            e = mid - 1
        else:
            s = mid + 1
            
    return s if nums[s]>nums[e] else e

def search(nums, target):
    n = len(nums)
    if n==1:
        if nums[0] == target:
            return True
        return False
    
    p = findPivot(nums)
    e = len(nums) - 1
    s = 0
    if p==target:
        return p
    if nums[s] == nums[e]:
        s += 1
        e -= 1
    if target>=nums[s]:
        return binarySearch(s,p-1,nums,target)
    else:
        return binarySearch(p+1, e, nums,target)
```
